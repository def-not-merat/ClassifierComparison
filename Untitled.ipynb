## Part (a):

* **Data Loading:** Read the lymph dataset from your mounted Google Drive using Pandas.

* **Data Splitting:** Split the dataset randomly into a 60% training set and a 40% test set using scikit-learn's train_test_split function.

* **Handling Categorical Data:** Encode categorical attributes with one-hot encoding using pandas.get_dummies, dropping the first attribute to avoid redundancy.

* **Decision Tree Classifier (1st Experiment):** Train a Decision Tree classifier in scikit-learn on the training data, setting the splitting criterion to "entropy," and requiring a minimum of 4 samples to split an internal node. Keep other settings unchanged.

* **Reproducibility:** Set the random_state to 32 in train_test_split and DecisionTreeClassifier to make your codes reproducible.

* **Performance Evaluation (1st Experiment):** Report the Decision Tree classifier's accuracy on the test data. Also, generate a high-resolution tree visualization (use plt.figure(figsize=(12, 8), dpi=300)).





```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
RNG = 42

# from google.colab import drive
# drive.mount('/content/drive')
#  df = pd.read_csv("/content/drive/MyDrive/data-mining-assignment3/datasets/lymph.csv")  # Load your dataset
df = pd.read_csv("datasets/lymph.csv")  # Load your dataset

#.............................................................................
# write the rest here
def one_hot(df, feature):
    feat = df[feature]
    df = df.drop(feature, axis='columns')
    dummies = pd.get_dummies(feat, prefix=feature[1:-1])
    dummies = dummies.drop(dummies.columns[0],axis='columns')
    merged = pd.concat([df, dummies], axis='columns')
    return merged
def yesno_to_binary(df, feature):
    df[feature] = df[feature].map({'yes':True ,'no':False})
    df = df.rename(columns={feature: feature[1:-1]})
    return df


features = df[df.columns[:-1]]
labels = df[df.columns[-1]]

for feat in df.columns[:-1]:
    if df[feat].iloc[0] in {'yes', 'no'}:
        features = yesno_to_binary(features, feat)
    else:
        features = one_hot(features, feat)


X_train, X_test, Y_train, Y_test = train_test_split(features, labels, train_size=0.6,random_state=RNG)

clf = DecisionTreeClassifier(criterion='entropy',
                            min_samples_split=4,
                            random_state=RNG)
clf = clf.fit(X_train, Y_train)

train_preds = clf.predict(X_train)
accuracy = accuracy_score(Y_train, train_preds)
print("accuracy on train: ", accuracy)

test_preds = clf.predict(X_test)
accuracy = accuracy_score(Y_test, test_preds)
print("accuracy on test: ", accuracy)

plt.figure(figsize=(12, 8), dpi=300)
_ = plot_tree(clf,
feature_names=list(features.columns),
class_names=sorted(labels.unique()))
#.............................................................................
```

##Part (a):
*  Write a Python function named "preprocess" that preprocesses and prepares
datasets for machine learning. It must take a training dataset and an optional test dataset as input. The code must first identify categorical and numerical attributes, then handle missing values by imputing the mean for numerical attributes and the most frequent value for categorical attributes. It is worth noting that to determine if an attribute is numerical, the process should involve checking its data type and potentially examining the number of unique values. This is necessary because an attribute may appear to have an "object" data type but could, in fact, be numerical in nature.

*  Subsequently, the "preprocess" code must standardize the numerical attributes by removing the mean and scaling to unit variance., and encodes categorical attributes using one-hot encoding.

*  If a test dataset is provided, it must undergo the same preprocessing steps, ensuring that the imputation for missing values and data scaling are performed based on the values obtained from the training dataset, thus maintaining consistency between the two datasets.

*  At the end, the function must return the preprocessed training and test datasets as (X_train, y_train, X_test, y_test) tuple if the test dataset exists. Otherwise it must return (X_train, y_train). X stands for independent processed attributes, while y indicates the class attribute.

```python
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
def preprocess(train_df, test_df=None):
    #.............................................................................
    # write your codes here
    numerics = []
    categoricals = []

    train_df = train_df.copy()
    # train_df = train_df[train_df[train_df.columns[-1]].notna()]

    y_train = train_df['class']

    train_df = train_df.loc[:, train_df.columns != 'class']

    for col in train_df.columns:
        to_num = pd.to_numeric(train_df[col], errors='coerce')
        if pd.api.types.is_numeric_dtype(train_df[col]) or to_num.notna().sum() > to_num.count()/2:
            numerics.append(col)
            train_df[col] = to_num
        else:
            categoricals.append(col)
            
    new_train = pd.DataFrame()
    
    if len(numerics) != 0:
        imp_mean = SimpleImputer(strategy='mean')
        imp_mean.set_output(transform='pandas')
        new_train[numerics] = imp_mean.fit_transform(train_df[numerics])

        scaler = StandardScaler()
        new_train[numerics] = scaler.fit_transform(new_train[numerics])

    if len(categoricals) != 0:
        imp_freq = SimpleImputer(missing_values='?', strategy='most_frequent')
        imp_freq.set_output(transform='pandas')
        X = imp_freq.fit_transform(train_df[categoricals])

        enc = OneHotEncoder(drop=None, sparse_output=False)
        encoded = enc.fit_transform(X)
        encoded = pd.DataFrame(encoded)
        encoded.columns = enc.get_feature_names_out()
        new_train = pd.concat([new_train, encoded], axis='columns')

    if test_df is not None:
        
        test_df = test_df.copy()
        # test_df = test_df[test_df[test_df.columns[-1]].notna()]
        
        y_test = test_df['class']

        test_df = test_df.loc[:, test_df.columns != 'class']

        new_test = pd.DataFrame()
        if len(numerics) != 0:
            for col in numerics:
                new_test[col] = pd.to_numeric(test_df[col], errors='coerce')
            new_test[numerics] = imp_mean.transform(new_test[numerics])
            new_test[numerics] = scaler.transform(new_test[numerics])
        if len(categoricals) != 0:
            X = imp_freq.transform(test_df[categoricals])
            encoded = enc.transform(X)
            encoded = pd.DataFrame(encoded)
            encoded.columns = enc.get_feature_names_out()
            new_test = pd.concat([new_test, encoded], axis='columns')

    #.............................................................................
    print("numerics:", len(numerics) ,"/", len(train_df.columns))
    if test_df is not None:
        return new_train, y_train, new_test, y_test
    else:
        # If no test dataset is provided, return only the preprocessed training dataset
        return new_train, y_train
```

```python
from math import floor
from sklearn.preprocessing import MinMaxScaler

class preprocess_fraud:
    def __init__(self):
        self.one_hot=None
        self.set_05=None
        self.set_mean=None
        self.linear_scaler=None
        self.scaler=None
    

    def fit_transform(self, fraud_dff):
        return self.transform(fraud_dff)

    def transform(self,fraud_dff):
        fraud_df = fraud_dff.copy()
        fraud_df = fraud_df.drop(["ORDER_ID","WEEKDAY_ORDER", "DATE_LORDER",'ANUMMER_01', 'ANUMMER_02', 'ANUMMER_03','ANUMMER_04', 'ANUMMER_05', 'ANUMMER_06', 'ANUMMER_07', 'ANUMMER_08','ANUMMER_09', 'ANUMMER_10'], axis='columns')
        fraud_df.replace({"no": 0, "yes": 1}, inplace=True)
        
        temp_bday = fraud_df.loc[fraud_df['B_BIRTHDATE'] != '?','B_BIRTHDATE'].apply(lambda x:2000 - int(str(x)[-4:]))
        fraud_df.loc[fraud_df['B_BIRTHDATE'] != '?','B_BIRTHDATE'] = temp_bday
        fraud_df.loc[fraud_df['B_BIRTHDATE'] == '?','B_BIRTHDATE'] = np.NaN

        temp_time = fraud_df.loc[fraud_df['TIME_ORDER'] != '?','TIME_ORDER'].apply(lambda x:int(x[:-3]))
        fraud_df.loc[fraud_df['TIME_ORDER']!='?','TIME_ORDER'] = temp_time
        fraud_df.loc[fraud_df['TIME_ORDER']=='?','TIME_ORDER'] = np.NaN


        categoricals = ['Z_METHODE','Z_CARD_ART','MAHN_AKT','MAHN_HOECHST']
        if self.one_hot is None:
            self.one_hot = OneHotEncoder(drop=None, sparse_output=False)
            encoded = self.one_hot.fit_transform(fraud_df[categoricals])
        else:
            encoded = self.one_hot.transform(fraud_df[categoricals])
        encoded = pd.DataFrame(encoded)
        encoded.columns = self.one_hot.get_feature_names_out()
        fraud_df = fraud_df.drop(categoricals, axis='columns')
        fraud_df = pd.concat([fraud_df, encoded], axis='columns')

        numeric_05 = ['Z_LAST_NAME']
        if self.set_05 is None:
            self.set_05 = SimpleImputer(strategy='constant', missing_values='?', fill_value=0.5)
            self.set_05.set_output(transform='pandas')
            fraud_df[numeric_05] = self.set_05.fit_transform(fraud_df[numeric_05])
        else:
            fraud_df[numeric_05] = self.set_05.transform(fraud_df[numeric_05])


        numeric_mean = ['B_BIRTHDATE','TIME_ORDER']
        if self.set_mean is None:
            self.set_mean = SimpleImputer()
            self.set_mean.set_output(transform='pandas')
            fraud_df[numeric_mean] = self.set_mean.fit_transform(fraud_df[numeric_mean])
        else:
            fraud_df[numeric_mean] = self.set_mean.transform(fraud_df[numeric_mean])


        fraud_df['Z_CARD_VALID'] = fraud_df['Z_CARD_VALID'].apply(lambda x:(x-floor(x)-0.2)*10000)


        needs_scaling = ['B_BIRTHDATE', 'VALUE_ORDER', 'AMOUNT_ORDER', 'SESSION_TIME', 'AMOUNT_ORDER_PRE', 'VALUE_ORDER_PRE']
        if self.scaler is None:
            self.scaler = StandardScaler()
            fraud_df[needs_scaling] = self.scaler.fit_transform(fraud_df[needs_scaling])
        else:
            fraud_df[needs_scaling] = self.scaler.transform(fraud_df[needs_scaling])

        needs_linear_scaling = ['Z_CARD_VALID', 'TIME_ORDER']
        if self.linear_scaler is None:
            self.linear_scaler = MinMaxScaler()
            fraud_df[needs_linear_scaling] = self.linear_scaler.fit_transform(fraud_df[needs_linear_scaling])
        else:
            fraud_df[needs_linear_scaling] = self.linear_scaler.transform(fraud_df[needs_linear_scaling])

        return fraud_df
```
